```python
spresto = pd.read_csv('data/spresto3_random.csv')
spresto.columns
```

```python
pd.set_option('display.max_columns', None)
spresto.head(20)
```

```python
col_name = ['subject_id', 'eczema_cum', 'rhinitis_cum', 'wheeze_cum', 'ethnicity',
       'delivery_age', 'pcv1_highest_edu', 'pcv1_household_income', 'pcv1_vig', 'pcv1_mvpa',
       'pgv3_vig', 'pgv3_mvpa', 'pgv3_currently_smoking', 'pcv1_passive_smoke',
       'pgv3_passive_smoke', 'delivery_mode', 'sex', 'ga', 'stai_state_ndv2',
       'stai_trait_ndv2', 'stai_state_m3', 'stai_trait_m3', 'bdi_tot_ndv2',
       'bdi_tot_m3', 'epds_tot_ndv2', 'epds_tot_m3', 'infection',
       'pcv1_ever_drink_alcohol', 'pgv3_alcohol_3m', 'parity', 'weight_birth',
       'antibiotics_labour', 'ever_antibiotics', 'bisq_q3_m6',
       'bisq_totslphrs_m6', 'bisq_q3_m12', 'bisq_totslphrs_m12', 'bf_m6',
       'maternal_allergy_pcv1', 'cat', 'dog', 'expressway']
```

```python
# reorder the columns to match the order of GUSTO dataset
spresto_ordered = spresto[col_name]
spresto_ordered.shape
```

```python
# mapping some categorical data to gusto
spresto_ordered['sex'] = np.where(spresto_ordered['sex']==1, 0, 
     np.where(spresto_ordered['sex']==2, 1, np.nan))
```

```python
# map highest education
condlist = [spresto_ordered['pcv1_highest_edu']==1,
spresto_ordered['pcv1_highest_edu']==2,
spresto_ordered['pcv1_highest_edu']==4,
spresto_ordered['pcv1_highest_edu'].isin([3, 6]),
spresto_ordered['pcv1_highest_edu'].isin([5, 7, 8])
]
choicelist = [2, 3, 4, 5, 6]
spresto_ordered['pcv1_highest_edu'] = np.select(condlist, choicelist, np.nan)
```

```python
# map household income
condlist = [spresto_ordered['pcv1_household_income']==1,
spresto_ordered['pcv1_household_income'].isin([2, 3]),
spresto_ordered['pcv1_household_income']==4,
spresto_ordered['pcv1_household_income'].isin([5, 6, 7, 8, 9, 10])
]
choicelist = [2, 3, 4, 5]
spresto_ordered['pcv1_household_income'] = np.select(condlist, choicelist, np.nan)
```
