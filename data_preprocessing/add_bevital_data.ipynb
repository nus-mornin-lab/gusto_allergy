```python
gusto_bevital.info()
```

```python
spresto_bevital = pd.read_csv('data/spresto_bevital.csv')
spresto_bevital.rename(columns=colname_spresto, inplace=True)
print(spresto_bevital.shape)
spresto_bevital.head()
```
